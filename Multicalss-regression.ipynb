## Cоздание и обучение модели

<span style="color:green">Отделям целевую переменную от признаков</span>

```python
X = df.drop('species', axis = 1)
y = df['species']
```

<span style="color:green">Разбиваем данные на обучающие и тестовые наборы</span>


```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

<span style="color:green">Масштабируем признаки</span>

```python
scaler = StandardScaler()
```

```python
scX_train = scaler.fit_transform(X_train)
```

```python
scX_test = scaler.transform(X_test)
```

<span style="color:green">Создаем модель</span>

<span style="color:blue">penalty</span> - ответственен за штрафы l1, l2 или elastic_net для регуляризации

В случае elastic_net появляется гиперпараметр l1_ratio.

<span style="color:blue">multiclass</span> - параметр ovr (One-vs-Rest) показывает, что мы сначала отделяем один параметр от всех остальных, далее последовательно отделяем каждый последующий признак от всех остальных, пока не останется один

<span style="color:blue">solver</span> - разные алгоритмы для решения задач оптимизации. Стоит отметить, что для каждой методологии мультикласс подходят только опредлеенные алгоритмы оптимизации

<span style="color:blue">C</span> -значение лямбда в формуле для штрафа, насколько сильно мы будем применять штрафное слагаемое (рекомендуются логарифмические значения)

```python
log_model = LogisticRegression(solver = 'saga',
                              multi_class= 'ovr',
                              max_iter = 5000)
```

Указываем параметры для перебора GridSearchCV

```python
#Задаем параметры
penalty = ['l1', 'l2', 'elasticnet']
l1_ratio = np.linspace(0, 1, 20)
C = np.logspace(0, 10, 20)

#Создаем сетку
param_grid = {'penalty': penalty,
             'l1_ratio': l1_ratio,
             'C': C}
```

```python
grid_model = GridSearchCV(log_model, param_grid = param_grid)
```

```python
grid_model.fit(scX_train, y_train)
```

<span style="color:green">Метрики оценки модели</span>

```python
grid_model.best_params_
```

Лучшей моделью оказалась L1, поэтому l1-ratio не используется

```python
y_pred = grid_model.predict(scX_test)
```

```python
y_pred
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
ConfusionMatrixDisplay.from_estimator(grid_model, scX_test, y_test);
```

Модель идеально отработала для всех случаев кроме одного: virginica был по ошибке классифицирован как versicolor

```python
print(classification_report(y_test, y_pred))
```

Для setosa все метрики идеальны, так как этот класс идеально отделяется от остальных.

```python
from sklearn.metrics import roc_curve, auc
def plot_multiclass_roc(clf, X_test, y_test, n_classes, figsize=(5,5)):
    y_score = clf.decision_function(X_test)

    # создаём пустые структуры
    fpr = dict()
    tpr = dict()
    roc_auc = dict()

    # один раз вычисляем dummies 
    y_test_dummies = pd.get_dummies(y_test, drop_first=False).values
    for i in range(n_classes):
        fpr[i], tpr[i], _ = roc_curve(y_test_dummies[:, i], y_score[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    # roc для каждого класса
    fig, ax = plt.subplots(figsize=figsize)
    ax.plot([0, 1], [0, 1], 'k--')
    ax.set_xlim([0.0, 1.0])
    ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title('Receiver operating characteristic example')
    for i in range(n_classes):
        ax.plot(fpr[i], tpr[i], label='ROC curve (area = %0.2f) for label %i' % (roc_auc[i], i))
    ax.legend(loc="best")
    ax.grid(alpha=.4)
    sns.despine()
    plt.show()
    
plot_multiclass_roc(grid_model, scX_test, y_test, n_classes=3, figsize=(16, 10))
```
